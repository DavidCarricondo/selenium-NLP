```python
from wordcloud import WordCloud, STOPWORDS
import json
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
with open('../OUTPUT/data.json') as file:
    data = file.read()
data = json.loads(data)    
```

```python
# join all documents in corpus
text = ' '.join([v for e,v in data.items()])
```

```python
mask = np.array(Image.open("../INPUT/WC_mask.png"))
```

```python
wc = WordCloud(
    background_color="white",
    stopwords=STOPWORDS,
    max_words=1000,
    mask=mask,
    max_font_size=90,
    random_state=42,
    contour_width=1,
    contour_color="#119DFF",
)
wc.generate(text)
```

```python
fig = plt.figure(figsize=[8, 8])
ax = plt.imshow(wc.recolor(), interpolation="bilinear")
plt.axis("off")
```
